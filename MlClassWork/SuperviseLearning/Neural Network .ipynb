```python
class ANN(object):
    def __init__(self):
        self.input_unit = 2
        self.hidden_layer = 1
        self.hidden_unit = 3
        
        
        self.w1 = np.random.randn(self.input_unit,self.hidden_unit)
        self.w2 = np.random.randn(self.hidden_unit,self.hidden_layer)

    def forward(self,X):
        x = np.array(X)
        self.z2 = np.dot(x,self.w1)
        self.a2 = self.sigmoed(self.z2)
        self.z3 = np.dot(self.a2,self.w2)
        self.yhat = self.sigmoed(self.z3)
        return self.yhat
        
    def sigmoed(self,z):
        return (1/(1+np.exp(-z)))
        
        
```
